# Define the simulation conditions

```python
from math import cos, radians

def average(num_1, num_2 = None):
    if type(num_1) is list:
        return sum(num_1) / len(num_1)
    else:
        if num_2 is not None:
            numbers = [num_1, num_2]
            average = sum(numbers) / len(numbers)
            return average
        else:
            return num_1

# chemical dimensions in Angstroms and degrees
chemical_dimensions = {
    'bond':{
        'c-c':average(1.2,1.54),  # https://en.wikipedia.org/wiki/Bond_length
        'c--o':1.36  # https://en.wikipedia.org/wiki/Carbon%E2%80%93oxygen_bond
    },
    'angle':{
        'sp3':109.5,
        'sp2':120
    }
}

# abbreviate dimensions
cc_bond = chemical_dimensions['bond']['c-c']
mostly_forward_sp2 = cos(radians(chemical_dimensions['angle']['sp2']-90))
less_forward_sp2 = cos(radians(180-chemical_dimensions['angle']['sp2']))
extension_sp3 = (2*cc_bond**2*(1-cos(
    radians(chemical_dimensions['angle']['sp3']))
                              ))**0.5 + (chemical_dimensions['bond']['c--o']*cos(radians(60)))

# calculate the total molecular length & width
width = (cc_bond*(4*mostly_forward_sp2 + 5*less_forward_sp2))
length = (4*cc_bond*(mostly_forward_sp2 + less_forward_sp2)) + extension_sp3
print(f'{width} x {length} Angstroms')
```

# Execute the simulation

```python
from scipy.constants import micro
from uncertainties import ufloat
from numpy import log10
from math import isclose, pi
%run ../../pdipy/pdi.py
%matplotlib inline

# execute the simulation
for lux in [1e4, 4e4, 6e4]:
    # determine the experimental result
    time = 0.5
    if lux == 1e4:
        log_reduction = -log10(1-0.45)       
    elif lux == 4e4:
        log_reduction = -log10(1-0.94)
    elif lux == 6e4:
        log_reduction = -log10(1-0.98)
        
    # execute the simulation
    pdi = PDI(
        solution_dimensions = {
            'area_sqr_cm':pi/4,
            'depth_cm': 46,
        },
        verbose = True, 
        jupyter = True
    )
    pdi.define_conditions(
        bacterial_specie = 'S_aureus', 
        bacterial_cfu_ml = 3E7,
#         biofilm = True,
        surface_system = True,
        photosensitizer = 'protoporphyrin IX', 
        photosensitizer_characteristics = {
            'formula': {
                'value': 'C34_H34_N4_O4'   
            },
            'dimensions':{
                'length (A)': length,
                'width (A)': width,
                'depth (A)': 4,
                'shape': 'rect',
            }
        },
        area_coverage = 0.6*0.06,
        light_source = 'incandescent',
        measurement = {
            'lux': lux
        }
    )
    pdi.simulate(
        export_name = f'Beirao_et_al, biofilm, {lux} Lux solution',
        figure_title = f'{lux} Lux, Protoporphyrin IX',
        experimental_data = {
            'x': [time],
            'y': [log_reduction]
        },
        display_fa_oxidation = True,
        export_contents = False
    )
    display(pdi.processed_data)

    value, unit = pdi.parse_data(
        log_reduction = log_reduction
    )            
    if value is not None:
        acceptable_range_float = ufloat(time, time/2)
        if isclose(value, time, abs_tol=time/2):
            print(f'The prediction of {value} is within the {acceptable_range_float} tolerance of the actual value.\nA {round(value-time, 3)/time*100}%-error')
        else:
            print(f'--> ERROR: The prediction of {value} is not within the {acceptable_range_float} tolerance of the actual value.\nA {round(value-time, 3)/time*100}%-error')
    
    print('\n\n\n\n')
```
