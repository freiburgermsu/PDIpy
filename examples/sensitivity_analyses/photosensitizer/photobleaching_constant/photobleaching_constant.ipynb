```python
%run ../../../../pdipy/pdi.py
from numpy import logspace
import pandas

pandas.set_option('max_rows', 10)

pdi = PDI(printing = False)

pb_constants = logspace(5, -2, num = 8)
for constant in pb_constants:
    name = f'{constant} constant'
    pdi.define_conditions(
        'S_aureus', 
        bacterial_cfu_ml = 1e5,
        photosensitizer = 'A3B_4Zn',
        photosensitizer_characteristics = {
            'photobleaching_constant (cm2/(J*M))': {
                'value': constant
            }
        },
        photosensitizer_molar = 18e-7,
        light_source = 'LED',
        measurement = {'lux':20000}
    )
    pdi.simulate(
        name, 
        figure_title = name,
#         exposure_axis = True,
        display_fa_oxidation = False,
        display_inactivation = False,
        display_ps_excitation = True,
        export_contents = True
    )
```
