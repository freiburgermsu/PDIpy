# Execute the simulation

```python
from numpy import log10, logspace
%run ../../../../pdipy/pdi.py
%matplotlib inline

# execute the simulation
for source in ['LED', 'incandescent', 'fluorescent']:
    name = f'{source}'
    pdi = PDI(verbose = False, printing = False, jupyter = True)
    
    pdi.define_conditions(
        photosensitizer_molar = 5e-6,
        light_source = source,
        measurement = {'irradiance': 4}
    )
    pdi.simulate(
        export_name = name,
        figure_title = name,
        display_fa_oxidation = True,
        display_inactivation = True,
        export_contents = True
    )
```
