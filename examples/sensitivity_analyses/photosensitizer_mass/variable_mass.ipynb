# Execute the simulations

```python
%run ../../../pdipy/pdi.py
%matplotlib inline

# execute the simulation
for carbons in range(50,120,10):
    name = f'{carbons} carbons'
    pdi = PDI(verbose = False, jupyter = True)
    pdi.define_conditions(
        bacterial_specie = 'S_aureus', 
        surface_system = True,
        photosensitizer_characteristics = {
            'formula': {
                "value": f"C{carbons}_H48_N16_F12_Zn"
            }
        },
        measurement = {'irradiance': 4}
    )
    pdi.simulate(
        export_name = name,
        figure_title = name,
        display_fa_oxidation = True,
        display_inactivation = True,
        export_contents = False
    )
```
