```python
%run ../../../pdipy/pdi.py
from numpy import logspace
import pandas

pandas.set_option('max_rows', 10)

pdi = PDI(total_time = 720)

pb_constants = logspace(-1, -5, num = 10)
for constant in pb_constants:
    print(constant)
    name = f'{constant} constant'
    pdi.define_conditions(
        'S_aureus', 
        bacterial_cfu_ml = 1e5,
        photosensitizer = 'A3B_4Zn',
        photosensitizer_characteristics = {
            'photobleaching_constant (cm^2/J)': {
                'value': constant
            }
        },
        photosensitizer_molar = 18e-9,
        light_source = 'LED',
        measurement = {'lux':20000}
    )
    pdi.simulate(
        name, 
        figure_title = name,
#         exposure_axis = True,
        display_fa_oxidation = True,
        display_ps_excitation = True,
        export_contents = False
    )
    print('watts', pdi.parameters['watts'])
    display(pdi.raw_data)
```
