# Execute the simulation

```python
from numpy import log10, logspace
%run ../../../../pdipy/pdi.py
%matplotlib inline

# execute the simulation
for cfu_ml in logspace(4,10,7):
    name = f'{log10(cfu_ml)}-log CFU-mL'
    pdi = PDI(verbose = False, jupyter = True)
    
    pdi.define_conditions(
        bacterial_specie = 'S_aureus', 
        bacterial_cfu_ml = cfu_ml,
#         biofilm = True,
        photosensitizer_molar = 5e-6,
        measurement = {'irradiance': 4}
    )
    pdi.simulate(
        export_name = name,
        figure_title = name,
        display_fa_oxidation = True,
        display_inactivation = True,
        export_contents = True
    )
```
