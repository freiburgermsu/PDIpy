```python
%run ../../../pdipy/pdi.py
from numpy import logspace
import pandas

pandas.set_option('max_rows', 10)

pdi = PDI(verbose = False, printing = False)

lux_intensities = logspace(1, 5, num = 10)
for lux in lux_intensities:
    name = f'{lux} lux'
    pdi.define_conditions(
        'S_aureus', 
        bacterial_cfu_ml = 1e5,
        photosensitizer = 'A3B_4Zn',
        photosensitizer_molar = 18e-9,
        light_source = 'LED',
        measurement = {'lux':lux}
    )
    pdi.simulate(
        name, 
        figure_title = name,
        exposure_axis = False,
        display_inactivation = False,
        display_ps_excitation = True,
        export_contents = True,
    )
#     print('watts', pdi.parameters['watts'])
    display(pdi.raw_data)
```
