```python
%run ../../../pdipy/pdi.py
from numpy import logspace
import pandas

pandas.set_option('max_rows', 10)

pdi = PDI(verbose = True, printing = True, jupyter = True)

lux_intensities = logspace(1, 5, num = 10)
for lux in lux_intensities:
    name = f'{lux} lux'
    pdi.define_conditions(
        'S_aureus', 
        biofilm = False,
        bacterial_cfu_ml = 1E8,
        photosensitizer = 'A3B_4Zn',
        photosensitizer_molar = 18e-7,
        light_source = 'LED',
        measurement = {'lux':lux}
    )
    try:
        pdi.simulate(
            name, 
            figure_title = name,
            exposure_axis = False,
            display_inactivation = True,
            display_fa_oxidation = True,
            display_ps_excitation = True,
            export_contents = False,
        )
#     print('watts', pdi.parameters['watts'])
    except:
        pass
    display(pdi.raw_data)
```
