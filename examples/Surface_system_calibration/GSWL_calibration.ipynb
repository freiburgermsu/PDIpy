# A3B_4Zn porphyrin photosensitizer dimensions

```python
from math import cos, radians

def average(num_1, num_2 = None):
    if type(num_1) is list:
        return sum(num_1) / len(num_1)
    else:
        if num_2 is not None:
            numbers = [num_1, num_2]
            average = sum(numbers) / len(numbers)
            return average
        else:
            return num_1
        
# chemical dimensions in Angstroms (as the averages from https://en.wikipedia.org/wiki/Bond_length) and degrees
chemical_dimensions = {
    'bond':{
        'c-c':average([1.2,1.54]),
        'c-h':average([1.06,1.12]),
        'c-n':average([1.47,2.1]),
        'c-f':average(1.34),
        'n=n':average(1.23) # https://doi.org/10.1016/B978-0-08-101033-4.00003-6
    },
    'angle':{
        'sp3':109.5,
        'sp2':120
    }
}

# abbreviate dimensions
cc_bond = chemical_dimensions['bond']['c-c']
cn_bond = chemical_dimensions['bond']['c-n']
cf_bond = chemical_dimensions['bond']['c-f']
mostly_forward_sp2 = cos(radians(chemical_dimensions['angle']['sp2']-90))
less_forward_sp2 = cos(radians(180-chemical_dimensions['angle']['sp2']))
mostly_forward_sp3 = cos(radians(chemical_dimensions['angle']['sp3']-90))
less_forward_sp3 = cos(radians(chemical_dimensions['angle']['sp3']-90))


# calculate the volume of conjugated region of the photosensitizer
center_porphyrin_length = cc_bond*(3*mostly_forward_sp2 + less_forward_sp2)
sp2_extension = cc_bond * (2 + less_forward_sp2) + cn_bond * (1 + less_forward_sp2)
sp3_diazirine = less_forward_sp3*(cc_bond*(3+less_forward_sp2+mostly_forward_sp2)) + cc_bond*mostly_forward_sp3 + cf_bond*mostly_forward_sp3

# calculate the total molecular length & width
conjugated_length = center_porphyrin_length + 2*sp2_extension
total_length = center_porphyrin_length + 2*(sp2_extension+sp3_diazirine)
linked_total_length = total_length - (cc_bond*mostly_forward_sp3 + cf_bond*mostly_forward_sp3)
print('conjugated_length', conjugated_length, 'Angstroms')
print('total_length', total_length, 'Angstroms')
print('linked_length', linked_total_length, 'Angstroms')
```
